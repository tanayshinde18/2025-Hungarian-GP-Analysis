```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv("../data/2025_Hungary_race_laps_with_stint.csv")
```

```python

pits_count=df[['Abbreviation', 'PitStopCount','DriverNumber']].drop_duplicates()
sns.barplot(data=pits_count, x='DriverNumber', y='PitStopCount', palette='viridis')
plt.title("Pit Stops per Driver")
# plt.figure(figsize=(50, 10))
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

for driver in df['Abbreviation'].unique():
    driver_data = df[df['Abbreviation'] == driver]
    ax.scatter(driver_data['LapNumber'], [driver]*len(driver_data),
               c=driver_data['Compound'].map({'SOFT': 'Green', 'MEDIUM': 'blue', 'HARD': 'red'}),
               edgecolor='black')

ax.set_title("Tyre Compound Usage Over Race")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Driver")
plt.show()
```

```python
stint_lengths = (
    df.groupby(['Abbreviation', 'StintNumber'])
    .size()
    .reset_index(name='LapCount')
)

sns.boxplot(data=stint_lengths, x='StintNumber', y='LapCount')
plt.title("Distribution of Stint Lengths")
plt.show()
```

```python
# Convert LapTime, PitInTime, PitOutTime to numeric seconds
time_cols = ['LapTime', 'PitInTime', 'PitOutTime']
for col in time_cols:
    df[col] = pd.to_timedelta(df[col], errors='coerce').dt.total_seconds()

# Convert Abbreviation to category
df['Abbreviation'] = df['Abbreviation'].astype('category')

# Check results
print(df.dtypes)
print(df.head())
```
